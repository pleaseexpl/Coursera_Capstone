```python
import pandas as pd
import numpy as np

toronto = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')# scrape data from Wikipedia
toronto_neighborhood = toronto[0]
print(toronto_neighborhood.shape)
toronto_neighborhood.head()
```

```python
isinstance(toronto_neighborhood, pd.DataFrame) # verifying have dataframe
```

```python
toronto_neighborhood_filtered = toronto_neighborhood[toronto_neighborhood['Borough']!= 'Not assigned']# removed Boroughs = 'not assigned'
```

```python
print(toronto_neighborhood_filtered.shape)
toronto_neighborhood_filtered.head()
```

```python
toronto_neighborhood_filtered = toronto_neighborhood_filtered.reset_index(drop=True) # reset the index
print(toronto_neighborhood_filtered.shape)
toronto_neighborhood_filtered
```
